## 缺失值补全

```python
# notRepairedDamage，汽车有尚未修复的损坏：是：0，否：1，但是数据中存在-，需要进行处理
train_data['notRepairedDamage'].value_counts()
# train_data['notRepairedDamage'].describe()
```

```python
# 不同背景不同填充方法，常见的就是均值、中位数、众数。或者根据业务背景填充，
# 比如说同一个车型下，车身信息有缺失，那你就找同一个车型有车身信息的来补充。或者是直接不填充，直接上lgb
train_data['notRepairedDamage'].replace('-', '0.0', inplace=True)
#train_data['notRepairedDamage'].mode()[0]
```

```python
test['notRepairedDamage'].value_counts()
test['notRepairedDamage'].replace('-', '0.0', inplace=True)
```

```python
# 对power异常值进行处理
# 缺失值用-1补全
train_data = train_data.fillna(-1)
test = test.fillna(-1)
```

```python
# 查看数值类型
train_data.info()
```

```python
# 查看数值类型
numerical_cols = train_data.select_dtypes(exclude='object').columns
numerical_cols
```

```python
# 查看分类类型
categorical_cols = train_data.select_dtypes(include='object').columns
categorical_cols
```

```python
# 特征选择
drop_cols = ['SaleID', 'regDate', 'creatDate', 'offerType', 'price']
feature_cols = [col for col in train_data.columns if col not in drop_cols]
feature_cols
```

```python
# 提取特征列
X_data = train_data[feature_cols]
Y_data = train_data['price']
X_test = test[feature_cols]
```

```python
# 定一个统计函数，用于统计某字段的特征
def show_stats(data):
    print('min: ', np.min(data))
    print('max: ', np.max(data))
    # ptp = max - min
    print('ptp: ', np.ptp(data))
    print('mean: ', np.mean(data))
    print('std: ', np.std(data))
    print('var: ', np.var(data))
# 查看price
show_stats(Y_data)
```

```python
import warnings
warnings.filterwarnings('ignore')
#X_data.info()
X_data['notRepairedDamage'] = X_data['notRepairedDamage'].astype('float64')
X_test['notRepairedDamage'] = X_test['notRepairedDamage'].astype('float64')
```

```python
import xgboost as xgb
# 创建模型
model = xgb.XGBRegressor(n_estimators=500, learning_rate=0.01, gamma=0, subsample=0.8, 
                         colsample_bytree=0.8, max_depth=7, random_state=2021)
model.fit(X_data, Y_data)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# 训练欠拟合，n_estimators太小或者learning_rate太小
show_stats(y_pred)
```
